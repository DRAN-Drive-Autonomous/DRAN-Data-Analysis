# Data Analysis

```python
print(f'Total Values: {len(df.index)}')
```

```python
throttle_count = df['Throttle Value'].value_counts()
```

```python
steering_count = df['Steering Value'].value_counts()
```

```python
print(throttle_count)
```

```python
throttle_count.sort_values().plot(kind='bar')
```

Throttle Values has almost equal for for positive throttle or no throttle but very less values for negative throttle.

```python
print(steering_count)
```

```python
steering_count.sort_values().plot(kind='bar')
```

Steering Values has almost equal for left steering or right steering but very large amount of values for no steering.
