# Imports

```python
import pandas
import numpy as np
```

# keys.csv

```python
df = pandas.read_csv("keys.csv", index_col=None)
```

```python
print(f'Total Values: {len(df.index)}')
```

```python
df.head(10)
```

```python
df['combined'] = df['Throttle Value'].astype(str) + '_' + df['Steering Value'].astype(str)
```

```python
unique_values_count = df['combined'].value_counts()
```

```python
unique_values_count.sort_values().plot(kind='bar')
```
